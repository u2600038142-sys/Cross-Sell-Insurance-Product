```python
#MVP

# Transaction

#CONFIG
from pyspark.sql import functions as F

src_path_csv = f"dbfs:/Volumes/cross_sell_insurance/00_landing/transaction/All_Policies.csv"
target_table = f'cross_sell_insurance.`01_feature_staging`.transaction_data'

#Ambil data dari (CATALOG:training.SCHEMA:00_landing.VOLUME:transaction) ke (CATALOG:training.SCHEMA:01_bronze.TABLE:transaction_data)

#Ingest VOLUME ke DF
df = (
    spark.read
    .option("header", True)
    .option("sep", ";")
    .option("inferSchema", True)
    .option("quote",'"')
    .option("escape",'"')
    .option("multiline", True)
    .option("mode", "PERMISSIVE")
    .csv(src_path_csv)
)

# Normalisasi nama kolom
df = df.toDF(*[c.strip().lower().replace(" ","_") for c in df.columns])

display(df)
```

```python
(
    df.write
    .mode("overwrite")
    .option("mergeSchema", "true")
    .format("delta")
    .saveAsTable(target_table)
)
```
