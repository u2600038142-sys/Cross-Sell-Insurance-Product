```python
from databricks.sdk import WorkspaceClient
import pandas as pd
import numpy as np

w = WorkspaceClient()

# 1) Ambil 1 row sample dari table
df_spark = spark.table("cross_sell_insurance.01_feature_staging.stage2_clean_feature_table").limit(1)
df = df_spark.toPandas()

# 2) Drop target kalau ada
X_sample = df.drop(columns=["is_target_customer"], errors="ignore")

# 3) BERSIHKAN NaN dan inf -> JSON friendly
X_sample = (
    X_sample
    .replace([np.inf, -np.inf], np.nan)  # ganti inf jadi NaN dulu
    .fillna(0)                           # semua NaN -> 0
)

# 4) Convert ke dict (records)
data = X_sample.to_dict(orient="records")

# 5) Panggil endpoint
result = w.serving_endpoints.query(
    name="Lapse_Score_Insurance1",   # ganti dengan nama endpoint kamu
    inputs=data
)

result
```
