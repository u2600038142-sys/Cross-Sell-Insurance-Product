```python
%sql
CREATE OR REPLACE TABLE cross_sell_insurance.02_golden.customer_360_prospect_bebas_aksi AS
SELECT 
    c360.*,
    scored.bebas_aksi_score,
    scored.bebas_aksi_flag
FROM cross_sell_insurance.01_feature_staging.stage3_bebas_aksi_scored scored
LEFT JOIN cross_sell_insurance.02_golden.customer_360 c360
    ON c360.client_id = scored.client_id
WHERE scored.bebas_aksi_flag = 1;
```
