```python
%sql
CREATE OR REPLACE TABLE cross_sell_insurance.01_feature_staging.stage1_early_feature_table AS

WITH target_customers AS (
    SELECT DISTINCT accountid
    FROM cross_sell_insurance.01_feature_staging.transaction_data
    WHERE accountid IN (
        SELECT accountid
        FROM cross_sell_insurance.01_feature_staging.transaction_data
        WHERE CNTTYPE_desc = 'E Comm PA'
    )
    AND CNTTYPE_desc <> 'E Comm PA'
)

SELECT
    c360.client_id,
    c360.birth_date,
    c360.insured_age,
    c360.gender,
    c360.marital_status,
    c360.black_list,
    c360.occupancy,
    c360.fatca,
    c360.isfatcacrs,
    c360.fatca_indicia,
    c360.le17,
    c360.`18-21`,
    c360.`22-27`,
    c360.`28-34`,
    c360.`35-39`,
    c360.`40-49`,
    c360.ge50,
    c360.policy_counts,
    c360.customer_vintage_in_month,
    c360.agency_counts,
    c360.ecomm_counts,
    c360.accident_counts,
    c360.cl_counts,
    c360.rpul_counts,
    c360.spul_counts,
    c360.traditional_counts,
    c360.credit_card_payment,
    c360.bank_transfer_payment,
    c360.auto_debet_payment,
    c360.yearly_payment,
    c360.halfyearly_payment,
    c360.quarterly_payment,
    c360.monthly_payment,
    c360.single_payment,
    c360.ape_sums,
    c360.inforce_counts,
    c360.lapse_counts,
    c360.fwd_max_flag,
    c360.automotive,
    c360.`book_&_movie`,
    c360.children,
    c360.culinary,
    c360.fashion,
    c360.gadget,
    c360.gasoline,
    c360.`go-pay`,
    c360.health,
    c360.home_expense,
    c360.music,
    c360.property,
    c360.shopping,
    c360.sport,
    c360.style,
    c360.travel,
    CASE WHEN t.accountid IS NOT NULL THEN 1 ELSE 0 END AS is_target_customer
FROM cross_sell_insurance.02_golden.customer_360 c360
LEFT JOIN target_customers t
    ON c360.client_id = t.accountid
ORDER BY c360.client_id;
```
